### Равномерное распределение
Абсолютно непрерывная случайная величина $X$ называется *равномерно распределённой* на отрезке $[a, b]$, если её плотность задаётся функцией 
$$p_{X}(x)=\frac{1}{b-a}\mathbb I_{[a, b]}(x),$$
где $\mathbb I_{[a, b]}$ — индикаторная функция отрезка $[a, b]$. Пишут:

$$\newcommand{\Uniform}{\mathop{\mathrm{Uniform}}}
X\sim \Uniform(0, 1)$$

### Задача 1 (5 баллов)
Пусть $X \sim \Uniform(-1, 2)$, $Y=X^2$.

Найти плотность $p_Y(y)$. Является ли она ограниченной?

### Задача 2 (15 баллов)
Пусть 

$$
\begin{gather*}
X  \sim \Uniform(-1, 1),\\
\varepsilon \sim \Uniform(0, 1),\\
Y=X+\varepsilon X,
\end{gather*}
$$
где $\varepsilon$ независим от $X$.

1. Найти $p_{X}(x)$ (плотность маргинального распределения $X$), $p_{Y\mid X=x}(y)$ (условную плотность для распределения $Y\mid X=x$ для заданного $x$), $p_{X, Y}(x, y)$ (совместную плотность).
2. Найти $\mathbb E[Y|X=x_0]$ (это какая-то функция от числа $x_0$).
3. Найти $p_{Y}(y)$ — маргинальную плотность $Y$. Проверьте, что это действительно плотность, то есть $\int_{\mathbb R}p_Y(y)dy=1$. Является ли эта функция ограниченной?
4. Как изменился бы ответ на вопрос 3, если бы $\varepsilon$ не был независим от $X$, а вычислялся по формуле: $\varepsilon = |X|$. (Проверьте, что в этом случае распределение $\varepsilon$ по-прежнему $\Uniform(0, 1)$.)

### Задача 3 (10 + 10 баллов)

Рассмотрим случайную величину $Y$, имеющую плотность $p(y)$, которую мы будем считать известной функцией. Пусть $p$ непрерывна; рассмотрим носитель меры, заданной плотностью $p$, то есть множество $\{y\mid p(y)>0\}$. Предположим, что носителем является связное подмножество прямой (быть может, вся прямая). 

Мы хотим подобрать такую величину $\hat y \in \mathbb R$, чтобы матожидание функции потерь $\mathbb E_{y\sim Y} L(y, \hat y)$ было минимальным. 

1. Пусть $L(y, \hat y)=|y-\hat y|$. Выразить оптимальное $\hat y$ через функцию $p$. Докажите, что оптимальное $\hat y$ определяется однозначно.
2. Пусть $L(y, \hat y)=c_{-} \cdot [y < \hat y] \cdot (\hat y - y) + c_{+} \cdot [y \ge \hat y] \cdot (y - \hat y)$, где выражение $[\langle\text{условие}\rangle]$ принимает значение 1, если условие верно, и 0, если неверно ([скобка Айверсона](https://ru.wikipedia.org/wiki/Скобка_Айверсона)), а $c_{-}$ и $c_{+}$ — некоторые положительные константы. Пусть известно, что для такой функции потерь оптимальное значение $\hat y$ является [$\alpha$-квантилем](https://ru.wikipedia.org/wiki/Квантиль#Определение) распределения $Y$. Найти, чему могут быть равны $c_{-}$ и $c_{+}$.
